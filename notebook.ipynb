```python
from macos_mcp.desktop.service import Desktop
from time import time,sleep
```

```python
desktop=Desktop()
```

```python
sleep(2)
desktop_state=desktop.get_state(use_vision=True)
```

```python
print(desktop_state.tree_state.interactive_elements_to_string())
```

```python
desktop_state.screenshot
```

```python
import io
img = desktop_state.screenshot
if img is not None:
    buf = io.BytesIO()
    img.save(buf, format="PNG")
    file_size = len(buf.getvalue())
    print(f"Dimensions: {img.size[0]}x{img.size[1]} (width x height)")
    print(f"File size: {file_size:,} bytes ({file_size / (1024*1024):.1f} MB)")
    img
else:
    print("Screenshot is None (screen capture failed - check Accessibility permissions)")
```
